```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Country-data.csv')
country_names = np.asarray(df['country'])
df.drop('country', axis=1, inplace=True)

df_columns = df.columns
```

```python
df.cov()
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
learn_scaler=StandardScaler().fit(df)
standard_df = learn_scaler.transform(df)
standard_df = pd.DataFrame(standard_df, columns = df_columns)
```

```python
standard_df.cov()
```

```python
standard_df.corr()
```

```python
standard_df.corr()[standard_df.corr() > 0.8]
```

```python
standard_df.corr()[standard_df.corr() < -0.8]
```

```python
ones_frame = np.ones(standard_df.corr().shape)
lower_triangle = np.tril(ones_frame, k=-1)
lower = standard_df.corr().where(lower_triangle.astype('bool'))
to_drop = []
for column in lower.columns: 
    if any(lower[column]  > 0.8 ):
        to_drop.append(column)
to_drop
```

```python
ones_frame = np.ones(standard_df.corr().shape)
lower_triangle = np.tril(ones_frame, k=-1)
lower = standard_df.corr().where(lower_triangle.astype('bool'))

for column in lower.columns: 
    if any(lower[column]  < -0.8 ):
        to_drop.append(column)
to_drop
```

```python
to_drop = np.unique(to_drop)
to_drop
```

```python
standard_df.drop(to_drop, axis=1, inplace = True)
standard_df
```

```python
from sklearn.cluster import KMeans
wcss = [] 
for i in range(2, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(standard_df)
    wcss.append(kmeans.inertia_)
```

```python

import matplotlib.pyplot as plt
plt.plot(range(2, 11), wcss, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('WCSS')
plt.title('The Elbow Method using WCSS')
plt.show()
```

```python
import sklearn.metrics
silhouette_scores = [] 
for i in range(2, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(standard_df) 
    silhouette_scores.append(sklearn.metrics.silhouette_score(standard_df, kmeans.labels_))
```

```python
plt.plot(range(2, 11), silhouette_scores, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('silhouette')
plt.title('The Elbow Method using silhouette')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
# Instantiate the clustering model and visualizer
model = KMeans(init='k-means++', random_state=0)
visualizer = KElbowVisualizer(
    model, k=(2,11), metric='distortion'
)

visualizer.fit(standard_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# Instantiate the clustering model and visualizer
model = KMeans(init='k-means++', random_state=0)
visualizer = KElbowVisualizer(
    model, k=(2,11), timings=False, metric='silhouette'
)

visualizer.fit(standard_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# Instantiate the clustering model and visualizer
model = KMeans(init='k-means++', random_state=0)
visualizer = KElbowVisualizer(
    model, k=(2,11), timings=False, metric='calinski_harabasz'
)

visualizer.fit(standard_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from yellowbrick.cluster import SilhouetteVisualizer

  
fig, ax = plt.subplots(3, 2, figsize=(15,8))
for i in [2, 3, 4, 5, 6, 7]:
    '''
    Create KMeans instances for different number of clusters
    '''
    km = KMeans(n_clusters=i, init='k-means++', random_state=0)
    q, mod = divmod(i, 2)
    '''
    Create SilhouetteVisualizer instance with KMeans instance
    Fit the visualizer
    '''
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(standard_df) 
    print('For n_cluster = ' + str(i) + 'The average silhouette_score is : ', visualizer.silhouette_score_)
```

```python
import sklearn.metrics
from sklearn.mixture import GaussianMixture

silhouette_scores = [] 
for i in range(2, 11): 
    gm = GaussianMixture(n_components=i, random_state=0)
    gm.fit(standard_df) 
    silhouette_scores.append(sklearn.metrics.silhouette_score(standard_df, gm.predict(standard_df)))
```

```python
plt.plot(range(2, 11), silhouette_scores, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('silhouette')
plt.title('The Elbow Method using silhouette')
plt.show()
```
